```python
DF = tagDF(df)
PF = DF[~DF.payload.isna()][DF.verbose.isna()]
print(len(PF), len(PF.payload.unique()))
RES = {}
for index, row in PF.iterrows():
    PL = PF.loc[index,"payload"]
    if PL not in RES.keys():
        RES[PL] = row["ID"] 
for key, value in RES.items(): 
    PF.loc[(PF.payload == key) & (PF.verbose.isna()), "verbose"] = "Packet "+str(value) 
PF
```
