# Testing if acq works

```python
probe = usbProbe.HealsonUP20()
```

```python
plt.figure(figsize=(10,8)) 
plt.imshow(np.sqrt(probe.getImages(n=2)[1]),cmap=plt.cm.bone, aspect='auto')
```

```python
while(0): # to correct, of course

    frame = np.sqrt(probe.getImages(n=2)[1])
    # Display the frame
    plt.figure(figsize=(5,5))
    plt.imshow(frame, aspect='auto',cmap=plt.cm.bone)

    plt.axis('off') 
    plt.title("Probe")
    plt.show() 
    display.clear_output(wait=True)
```

# Exploring readable registers

```python
HEADER = "Unit"
for k in range(16):
    HEADER+= " "+str(k).rjust(4, ' ')
print(HEADER)
for k in range(16):
    LIGNE = str(k).rjust(4, ' ')
    for j in range(16):
        RES = probe.checkAddress(bytearray([16*k+j]))[0]
        LIGNE+= " "+str(RES).rjust(4, ' ')
    print(LIGNE)
```
