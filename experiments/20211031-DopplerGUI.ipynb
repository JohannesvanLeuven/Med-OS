```python
# import the necessary packages
from __future__ import print_function
from PIL import Image
from PIL import ImageTk
import tkinter as tki
import threading
import datetime
import imutils
import cv2 
import os
import time
import numpy as np  

import matplotlib.pyplot as plt 
```

```python
file = "data/20211030-npz/2021-10-30_18-37-16.npz"
filez = np.load(file)
filez.files
```

# Checking BMode image content

```python
RF = filez["arr_0"]

img = np.sqrt(np.sqrt(np.abs(RF[:-2])).T[7:].T)
img = np.array(255*(img/np.max(img))).astype(np.uint8)
print(img) 
img = cv2.merge([img,img,img])
img = cv2.resize(img, (400,400), interpolation = cv2.INTER_AREA)

plt.imshow(img,cmap= "bone")

```

# Checking Doppler image content

```python
DOP = filez["arr_1"]
DOP
```

```python
imgD = 128*(DOP/np.max(DOP))
imgD = cv2.applyColorMap(imgD.astype(np.uint8), cv2.COLORMAP_JET)
imgD = cv2.resize(imgD, (400,400), interpolation = cv2.INTER_AREA)

plt.imshow(imgD,cmap= "jet")
```

# Transforming the image

```python
rows,cols,channels = imgD.shape;
pts1 = np.float32([[0,0],[0,400],[400,0]]);
pts2 = np.float32([[170,75],[30+20,220],[370-20,75]]);
M = cv2.getAffineTransform(pts1,pts2);
imgD_transformed = cv2.warpAffine(imgD,M,(cols,rows));
plt.imshow(imgD_transformed,cmap= "jet")
```

```python
print(np.shape(img)),print(np.shape(imgD_transformed))
```

```python
overlay = cv2.addWeighted(img, 0.8, imgD_transformed, 0.8, 0.0)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(overlay,cmap= "jet")
```

```python
IMG1 = DOP[0::4]
plt.imshow(IMG1)
```

```python
imgD1 = 128*(IMG1/np.max(np.abs(IMG1)))
imgD1 = cv2.applyColorMap(imgD1.astype(np.uint8), cv2.COLORMAP_JET)
imgD1 = cv2.resize(imgD1, (400,400), interpolation = cv2.INTER_AREA)


rows,cols,channels = imgD1.shape;
pts1 = np.float32([[0,0],[0,400],[400,0]]);
pts2 = np.float32([[170,75],[30+20,220],[370-20,75]]);
M = cv2.getAffineTransform(pts1,pts2);
imgD1_transformed = cv2.warpAffine(imgD1,M,(cols,rows));
plt.imshow(imgD1_transformed,cmap= "jet")
```

### Adding thresholds

```python
RET = cv2.cvtColor(imgD1_transformed, cv2.COLOR_BGR2GRAY)
ret,Otsu = cv2.threshold(RET,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(Otsu)
fg = cv2.bitwise_or(imgD1_transformed, imgD1_transformed, mask=Otsu)
plt.imshow(fg)
```

```python
overlay = cv2.addWeighted(img, 0.8, fg, 0.8, 0.0)
plt.figure(figsize=(15,15))
plt.imshow(overlay,cmap= "jet")
```
